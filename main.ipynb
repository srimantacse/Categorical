```python
import pandas as pd
import math
import sys
import cv2
from numpy import array
from time import gmtime, strftime
from bigc import bigc
from index import dunn
from cvutils.io import imshow, imwrite
from fuzzy_kmodes import fKModes
import numpy as np

"""
############################# Data Set Parameters #############################
0: DONOT drop  | 1: drop Last      | 2: drop    First   | 3: drop First & Last      | 4: Special Case, First two
0: No Target   | 1: Target at Last | 2: Target at First | 3: Id First & Target Last | 4: Id First & Target second
"""
dataSetArray = [
                 #name                  noOfPoint   dropInfo noOfClass   P      separator  index   opt-delta
                ['mushroom.csv',        8124,       2,       2,          1000,  ','],      #0
                ['car.csv',             205,        0,       -1,         205,   ','],      #1
                ]

dataIndex = 1
dataSet   = 'data/'+ dataSetArray[dataIndex][0]
noOfPoints= dataSetArray[dataIndex][1]
drop      = dataSetArray[dataIndex][2]
noOfClass = dataSetArray[dataIndex][3]
P         = dataSetArray[dataIndex][4]
separator = dataSetArray[dataIndex][5]
delta     = .7

df_data = pd.read_csv(dataSet, sep=separator)

"""
#print(df_data.head(2))
#print(df_data.dtypes)
#print(df_data.sample(10))
#print(df_data.info())
#print(df_data.describe())
#print(df_data['bruises'].max())
#df_data.boxplot(column=['bruises'], figsize=(5,6))
#df_data_cat = df_data.select_dtypes(include=['object'])
#print(df_data_cat.head())
#print(df_data.isnull().sum())
"""

# Clean the result
count_nan1 = np.sum(len(df_data) - df_data.count())
for column in df_data.columns:
    df_data[column].fillna(df_data[column].mode()[0], inplace=True)
    
count_nan2 = np.sum(len(df_data) - df_data.count())
print("Cleaning Count: " + str(count_nan1 - count_nan2))

df_data_cat = df_data.select_dtypes(include=['object'])
df_data_cat = pd.get_dummies(df_data_cat, columns=['fuel-type'], drop_first=True)

print(df_data.sample(10))

dataX = None
dataY = None
if noOfClass != -1:
    dataY = array(df_data['class']).reshape(noOfPoints, 1)

if drop == 1:
    dataX = df_data.iloc[:, :-1]
elif drop == 2:
    dataX = df_data.iloc[:, 1:]
elif drop == 3:
    dataX = df_data.iloc[:, 1:-1]
elif drop == 4:
    dataX = df_data.iloc[:, 2:]

dataX = array(dataX)
print("Info:: Original Data   Size: " + str(dataX.shape))
if dataY != None:
    print("Info:: Original Class  Size: " + str(dataY.shape))
```
